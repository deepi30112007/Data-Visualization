```python
data={'RollNo':[6,7,4],
     'Name':['Apara','purvi','Bani'],
     'Marks':[92,90,98]
     }
df=pd.DataFrame(data)
print(df)
```

```python
df['Marks']=df['Marks']+5
print(df)
```

```python
df['Grade']=['A','A+','B']
print(df)
```

```python
df.drop('Grade',axis=1,inplace=True)
print(df)
```

```python
print(df['Marks'].sum())
```

```python
print(df['Marks'].mean())
```

```python
print(df['Marks'].max())
```

```python
print(df['Marks'].min())
```

```python
df.sort_values('Marks')
```
